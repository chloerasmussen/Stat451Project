```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv("healthcare-dataset-stroke-data.csv")
df = df.dropna().drop('id', axis = 1)
df = pd.get_dummies(df, columns=['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status'])
drops = ['gender_Male', 'gender_Other', 'ever_married_No', 'smoking_status_Unknown', 'work_type_children', 'Residence_type_Urban']
df = df.drop(drops, axis = 1)
df = df.rename(columns = {'gender_Female': 'Female', 'ever_married_Yes' : 'Married', 'work_type_Govt_job': 'Govt_Employee', 'work_type_Never_worked': 'Unemployed', 'work_type_Private' : 'Pvt_Employee', 'work_type_Self-employed': 'Self_Employed'})
df.head()
```
